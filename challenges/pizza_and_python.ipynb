# Challenge 3: Down with `for` loops

The code below uses `for` loops to perform a certain operation on arrays, but python for loops are inefficient for long computations. Your job is to rewrite these function using only Numpy array manipulations and library functions (e.g. `np.xxx()`). Do not use any `for` or `while` loops, iterators, generators, or list comprehensions in your solutions.

Given a vector $v$ of length $n$, and an integer $0<k<n$, return a 2D array as:

```
[[v[0], ..., v[k-1]],
 [v[1], ..., v[k]  ],
 [v[2], ..., v[k+1]],
 [ .     .       . ],
 [ .     .       . ],
 [ .     .       . ],
 [v[n-k+1, ..., v[n]]
 ```

> hint: https://numpy.org/doc/stable/user/basics.broadcasting.html



````python
# Challenge 3 Goals
import numpy as np
def sliding_stack_py(v, k):
    "Stack sliding windows of v of length k."
    rows = []
    for i in range(len(v) - k + 1):
        rows.append(v[i : (i + k)])
    return np.array(rows)

sliding_stack_py(np.array([1, 2, 3, 4, 5]), 3)
````
